```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Auto_price.csv")
```

```python
#bar graph
sns.countplot(x='symboling', data=df)
```

```python
sns.countplot(x='symboling', data=df, hue='aspiration')
```

```python
#line chart
sns.lineplot(x='stroke', y='bore', data=df)
```

```python
#scatter plot
df.plot.scatter('bore','stroke')
```

```python
df.plot.scatter('engine-size', 'symboling')
```

```python
#histogram
df['symboling'].hist(bins=8)
```

```python
df['stroke'].hist(bins=80)
```

```python
#pie chart
sizes = df['fuel-type'].value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=['gas', 'diesel'], autopct="%1.1f%%", shadow=True)
plt.show()
```
